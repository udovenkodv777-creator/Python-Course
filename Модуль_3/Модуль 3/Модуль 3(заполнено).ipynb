# <a id="for">1. Цикл for </a>

> **for переменная_цикла in объект:** \
&nbsp;&nbsp;&nbsp;&nbsp;инструкция \
&nbsp;&nbsp;&nbsp;&nbsp;... \
&nbsp;&nbsp;&nbsp;&nbsp;инструкция

Представим такую ситуацию: вы решили купить новый телевизор за 349 000 рублей. Но отдать за покупку разом столько денег вам жалко. Вы рассматриваете вариант рассрочки и хотите понять, на сколько месяцев надо брать рассрочку, чтобы и размер выплаты был комфортным, и чтобы не затягивать надолго с оплатой телевизора. Чтобы прикинуть, вы рассчитываете в Python, какой будет размер месячной оплаты при рассрочке на 3 месяца, 6, 9 и 12.

```python
price = 349000
periods = [3, 6, 9, 12]

for months in periods:
    payment = price / months
    print(f'На {months} мес: {payment:.2f} руб/мес')
```

```python
price = 349_000
for months in 3, 6, 9, 12:
    print('При рассрочке на', months, 'мес. - ', price/months)
    print('Недороговато ли?')
print('Теперь есть из чего выбрать')
```

# <a id="for-range">2. For и range </a>

Мы посмотрели на эти 4 варианта и решили, что нам этого мало: мы хотим увидеть все варианты, для любого количества месяцев – от минимума до 2 лет. Но печатать все эти значения мы не хотим, для этого – будем использовать функцию range(). \
Функция range() принимает 3 аргумента:
1. число, с которого необходимо начать
2. число, на котором необходимо остановиться (важно! само число, которое мы укажем, уже не будет входить в диапазон, поэтому указываем n+1) 
3. шаг, с которым будут увеличиваться числа (* если шаг равен 1, можно его не прописывать)

> **for переменная_цикла in range([start=0], stop, [step=1]):** \
&nbsp;&nbsp;&nbsp;&nbsp;инструкция \
&nbsp;&nbsp;&nbsp;&nbsp;... \
&nbsp;&nbsp;&nbsp;&nbsp;инструкция

```python
price = 349_000
for months in range(2, 13, 2):
    print('При рассрочке на', months, 'мес. - ', price/months)
    print('Недороговато ли?')
print('Теперь есть из чего выбрать')
```

Рассчитаем вариант для поквартальной рассрочки до пяти лет. Начало 3 месяца, верхняя граница количество месяцов * 5 +1, шаг 3 месяца - 1 квартал:

```python
price = 349_000
for months in range(3, 12 * 5 + 1, 3):
    print('При рассрочке на', months, 'мес. - ', price/months)
    print('Недороговато ли?')
print('Теперь есть из чего выбрать')
```

# <a id="range">3. Функция range() </a>

Если просто запутсить функцию `range()` отдельно, мы не увидим на выходе последовательность. \
Вместо неё – просто условия заданного диапазона.

```python
range(2, 12, 2)
```

Если напечатать – то же самое:

```python
print(range(2, 12, 2))
```

Если запросить тип – увидите тип range:

```python
type(range(2, 12, 2))
```

Чтобы потребовать от функции каждое из чисел и получить его, понадобится цикл. 

```python
for i in range(5, 16, 5):
    print(i)
```

```python
for i in range(2, 12):
    print(i)
```

```python
for i in range(6):
    print(i)
```

Вернемся к нашей задачке. Представим, что вы решили не покупать телевизор, слишком дорого, лучше вы пока что будете копить деньги. И понесли накопленния 349000 руб. в банк, положить под проценты. \
Например, на 5 лет, под 9 % годовых с выплатой раз в год. Вы хотите узнать, сколько это будет через пять лет. Но поскольку вы не собираетесь забирать из банка доход от процентов, а будете оставлять его на счету, вам нужно посчитать сложные проценты за пять лет - процент от основного вклада и от его процентов.

```python
sum = 349_000
for i in range(5):
    sum = sum * 1.09
    print('sum за', i+1, '-й год = ', round(sum,2))
```

Рассмотрим еще один пример. \
За окном осень. Ночи становятся всё длиннее, а температура на улице всё ниже. Можем приблизительно установить, что с каждой неделей ночь становится длиннее на полчаса, а температура ниже на полтора градуса. Давайте посчитаем, как она будет меняться по неделям в ближайшие 10 недель.


```python
night_length = 12.0  # Начальная длина ночи в часах
temperature = 15.0   # Начальная температура в градусах

for week in range(1, 11):
    night_length += 0.5
    temperature -= 1.5
    print(f"Неделя {week}: Ночь {night_length} ч, Температура {temperature} °C")
```

Есть более короткая запись, когда в цикле мы высчитываем значение переменной, исходя из её прошлого значения. Можно поставить оператор += или -= и указать значение, на которое хотим увеличить или уменьшить переменную

```python

sum = 349_000
for i in range(5):
    sum *= 1.09
    print(f'{i+1} год sum =  {sum:.2f}')
```

Кроме += и -= мы также можем использовать *= (когда вычисляем переменную, умножаем её же на какое-то число) или /= (деление равно).

# <a id="while">4. Цикл while </a>

> **while условие:** \
&nbsp;&nbsp;&nbsp;&nbsp;инструкция \
&nbsp;&nbsp;&nbsp;&nbsp;... \
&nbsp;&nbsp;&nbsp;&nbsp;инструкция

**Цикл while («пока») позволяет выполнить одну и ту же последовательность действий, пока проверяемое условие истинно.**

Как правило, цикл while используется, когда невозможно определить точное значение количества проходов исполнения цикла.

Вернемся к примеру про банковский вклад. Предположим, что мы не хотим создавать вклад на какое-то определенное количество лет, а просто хотим из наших 350 тысяч получить 700 тыс.руб. Когда эта сумма накопится, мы забираем деньги из банка. 

```python
money = 349_000
while money < 700_000:
    money *= 1.09 #money = money * 1.09
    print('На счету будет', money, 'рублей')
print('Забираю деньги')
```

```python
money_new = 14_000_000
while money_new < 50_000_000:
    money_new *= 1.145 #money_new = money_new * 1.145
    print ('На счету будет', round(money_new,2), "hруб.")
```

Мы можем добавить условие, при котором цикл прервётся в любом случае. \
Например, ограничить сумму в банке той, что может быть возмещена по страховке – сейчас это 1 миллион 400 тысяч. \
Если количество денег становится больше, мы останавливаем цикл. Добавим это условие. Выведем на печать намерение остановить цикл и добавим команду break – прерывание.

```python
money = 349_000
while money > 70_000:
    money *= 1.09 #money = money * 1.09
    print('На счету будет', money, 'рублей')
    if money >= 1_400_000:
        break
print('Забираю деньги')
```
