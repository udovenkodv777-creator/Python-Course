# Практика цикл for

1. С помощью цикла выведите на печать все числа от 5 до 30.

```python
for i in range(5, 31):
    print(i)
```

2. Напишите цикл, который выводит на печать числа, если они делятся на четыре без остатка. Расчёт должен начинаться с 5 и заканчиваться на 30 включительно.

5. Вводится строка text, состоящая из цифр. Рассчитать сумму цифр, входящих в строку. Если строка пустая, то вывести на экран 0. \
'123' → 6 \
'22' → 4 \
'1000' → 1   

```python
text = input("Введите строку из цифр: ")
# Превращаем каждый символ в число и складываем их
digit_sum = sum(int(char) for char in text)
print('Вы ввели',text)
print(f'Сумма цифр, входящих в введенную строку {text} составляет = {digit_sum}')


```

```python
text = input("Введите строку из цифр: ")
# Создаем переменную-аккумулятор для суммы
total_sum = 0
# Проходим по каждому символу в строке
for char in text:
    # Превращаем символ в число и прибавляем к общей сумме
    total_sum += int(char)
print(total_sum)
```

# Практика цикл while

1. По последним данным, население Торжка в 2020 году составляло 44 439 человек. Постепенно количество местных жителей снижается. Предположим, что с каждым годом население сокращается на 650 человек. Напишите цикл, который выводил бы на печать год и прогнозируемое число жителей Торжка, пока население не уменьшится до 40 тысяч человек.

```python
current_year = 2020
people = 44_439
dicline_per_year = 650
print(f'Количество жителей Торжка в {current_year} году = {people} человек')
while people > 40_000:
    people -= dicline_per_year
    current_year += 1
    print(f'Количество жителей Торжка в {current_year} году = {people} человек')
```

```python
population = 44439
year = 2020

while population >= 40000:
    print(f"{year} год — {population} чел.")
    population -= 650
    year += 1
```

2. При помощи цикла `while`, напишите программу, которая просит ввести любое число и выводит в ответ его квадрат. Программа должна повторять это действие бесконечно. Но, если человек вводит число 5, программа выводит на печать «дурак опять» и прекращает работать.

```python
while True:
    number = int(input("Введите любое число: "))
    
    if number == 5:
        print("дурак опять")
        break
    
    print(f"Квадрат числа: {number**2}")
```

3. Напишите цикл с использованием `while` – `else`, стартовой точкой которого было бы 25. В каждом прохождении цикла число должно умножаться на 3 и выводиться на печать. Когда число превысит один миллион, цикл должен остановиться, а на печать нужно вывести "стоп".

```python
s_number = 25
while s_number <= 1_000_000:
    s_number *= 3
    print(s_number)
else:
    print("стоп")
    
```
