Задача 2. Напишите программу, которая бесконечно запрашивает на ввод целое число и после каждого ввода печатает на экране ответ в соответствии со следующими условиями:
  * Если введённое число делится на три, то печатается это же число.
  * Если делится на пять, то на печать выводится его квадрат. Если же это число одновременно делится и на пять, и на три, то на экране должно появиться само число в соответствии с прошлым условием, а затем и его квадрат. То есть сначала должно отработать прошлое условие, а затем текущее. 
  * Если число делится на семь, нужно вывести на печать «дурак совсем» и прервать выполнение программы. 
  * Во всех остальных случаях программа должна снова запросить число на ввод.

`Подсказка`

Чтобы проверить, что программа работает верно, можно использовать следующие числа для теста: 
* 3 (в ответ печатается 3);
* 5 (в ответ печатается 25);
* 15 (в ответ печатается 15 и 225);
* 1 (ничего не печатается, запрашивается следующее число на ввод);
* 7 (в ответ печатается «дурак совсем», программа прерывается).

```python
# Ваш код
while True:
    num = int(input("Введите целое число: "))
    
    
    # проверка на условие, при котором завершается задание
    if num % 7 == 0:
        print("дурак совсем")
        break
        
    # проверка остальных условий по заданию
    if num % 3 == 0 and num % 5 == 0:
        print(f"{num} делится на 3 и на 5, его квадрат = {num ** 2}")
    elif num % 3 == 0:
        print(f"{num} делится на 3")
    elif num % 5 == 0:
        print(f"{num} делится на 5 и его квадрат {num ** 2}")
```

# Списки

Задача 3. Дан список `list1`, который содержит целые числа. Всего в списке 12 элементов. 

- Ваша задача — посчитать сумму элементов с нечётными индексами. Другими словами, нужно сложить элементы `list1[1]`, `list1[3]`, `list1[5]`, `list1[7]`, `list1[9]` и `list1[11]`. 

- Чтобы решить эту задачу, используйте slicing. 

`Подсказка`: вам придётся указать одну границу среза и шаг.   

```python
list1 = [1, 3, 0, -2, 4, 5, 9, 12, -9, 16, -75, 8]

# Ваш код здесь
# 1. Делаем срез элементов с нечётными индексами
slice1 = list1[1::2]

# 2. Считаем сумму элементов в полученном срезе
result_sum = sum(slice1)

# Выводим результаты для проверки
print("Элементы с нечётными индексами:", slice1)
print("Их сумма:", result_sum)
```

# Нормализация строк и функции

Задача 4. Очищение строк

Данные по фамилии и имени каждого пользователя приходят в виде строки вида: `Фамилия:Иванов,Имя:Герасим`. Напишите функцию, которая будет обрабатывать такую строку в качестве входного параметра, а на выходе будет возвращать два значения: фамилию и имя в нижнем регистре.

Проверьте, что ячейки ниже отработали верно.

```python
def get_name_data(input_data: str):
    # Ваш код здесь
     # Разделяем строку на две части по запятой
    last_part, first_part = input_data.split(",")
    # Убираем подстроку 'Фамилия:', очищаем от пробелов по краям и переводим в нижний регистр
    last_name = last_part.replace("Фамилия:", "").strip().lower()
    
    # Убираем подстроку 'Имя:', очищаем от пробелов по краям и переводим в нижний регистр
    first_name = first_part.replace("Имя:", "").strip().lower()
    
    # Возвращаем кортеж из двух значений
    return last_name, first_name
```

```python
get_name_data('Фамилия:Удовенко,Имя:Максим')
```

```python
get_name_data('Фамилия:Сидоров-Ильин,Имя:Петр')
```
