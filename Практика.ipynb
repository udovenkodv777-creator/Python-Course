```python
import pandas as pd

# 1. Читаем файл Excel (по умолчанию берется первый лист)
df = pd.read_excel('Таблица 1.xlsx')

# 2. Добавляем новую колонку: считаем выручку по каждому товару
df['Выручка'] = df['Цена'] * df['Количество']

# 3. Фильтруем: оставляем только крупные продажи (выручка больше 20 000)
large_sales = df[df['Выручка'] > 20000]

# 4. Сохраняем результат в новый файл Excel без вывода номеров строк (index=False)
large_sales.to_excel('large_sales_Таблица 1.xlsx', index=False)
```
