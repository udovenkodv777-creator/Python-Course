```python
#Симулятор рулетки на Python
import random

# Создаем список всех возможных исходов на европейской рулетке
# (число и его цвет в виде кортежа)
roulette_numbers = [("0", "зеленый")] + \
                   [(str(x), "красный") for x in [1,3,5,7,9,12,14,16,18,19,21,23,25,27,30,32,34,36]] + \
                   [(str(x), "черный") for x in [2,4,6,8,10,11,13,15,17,20,22,24,26,28,29,31,33,35]]

# Крутим рулетку!
winning_sector = random.choice(roulette_numbers)

number, color = winning_sector
print(f"Шарик упал на: {number} ({color})")
```

```python
#Программа принимает ставку через поле ввода, крутит рулетку и считает баланс игрока
import random

# Стартовый баланс монет
balance = 1000

# Множества для проверки цветов
red_numbers = {1, 3, 5, 7, 9, 12, 14, 16, 18, 19, 21, 23, 25, 27, 30, 32, 34, 36}
black_numbers = {2, 4, 6, 8, 10, 11, 13, 15, 17, 20, 22, 24, 26, 28, 29, 31, 33, 35}

print("🎰 Добро пожаловать в Python-рулетку!")
print("Вы можете ставить на: 'красное', 'черное', на конкретное число (0-36) или написать 'выход' для завершения игры.\n")

# Главный бесконечный цикл игры
while balance > 0:
    print("-" * 50)
    print(f"💰 Ваш текущий баланс: {balance} монет.")
    
    # 1. Принимаем ставку от пользователя
    bet_type = input("На что ставим? (красное / черное / число / выход): ").strip().lower()
    
    # Проверяем, хочет ли игрок выйти
    if bet_type == "выход":
        print(f"\n👋 Спасибо за игру! Вы уносите с собой {balance} монет.")
        break
        
    # Проверяем корректность типа ставки
    if bet_type not in ["красное", "черное"] and not (bet_type.isdigit() and 0 <= int(bet_type) <= 36):
        print("❌ Неверный тип ставки! Попробуйте еще раз.")
        continue

    # 2. Принимаем сумму ставки
    bet_amount_str = input(f"Сколько монет ставим? (1-{balance}): ").strip()
    if not bet_amount_str.isdigit():
        print("❌ Пожалуйста, введите число монет!")
        continue
        
    bet_amount = int(bet_amount_str)

    # Проверяем, хватает ли денег
    if bet_amount > balance or bet_amount <= 0:
        print(f"❌ Недостаточно монет или неверная сумма! У вас есть только {balance}.")
        continue

    # 3. Крутим рулетку
    winning_number = random.randint(0, 36)
    
    if winning_number == 0:
        winning_color = "зеленый"
    elif winning_number in red_numbers:
        winning_color = "красный"
    else:
        winning_color = "черный"
        
    print(f"\n🔮 Шарик крутится... и выпадает: {winning_number} ({winning_color})!")

    # 4. Проверяем логику выигрыша
    is_win = False
    payout_multiplier = 0

    if bet_type == "красное" and winning_color == "красный":
        is_win = True
        payout_multiplier = 2
    elif bet_type == "черное" and winning_color == "черный":
        is_win = True
        payout_multiplier = 2
    elif bet_type.isdigit() and int(bet_type) == winning_number:
        is_win = True
        payout_multiplier = 36

    # 5. Считаем итоги раунда
    if is_win:
        win_coins = bet_amount * (payout_multiplier - 1)
        balance += win_coins
        print(f"🎉 Вы выиграли {win_coins} монет!")
    else:
        balance -= bet_amount
        print(f"😢 Вы проиграли {bet_amount} монет.")

# Этот код сработает, когда баланс станет равен 0 и цикл прервется
if balance <= 0:
    print("-" * 50)
    print("💸 Упс! Ваши монеты закончились. Игра окончена!")
```

**Задача 2. Случайный выбор из списка** \
Дан список цветов, случайным образом выбрать и вывести один из них. \
`random.choice(sequence)` - случайный элемент непустой последовательности.

```python
colors = ['красный', 'синий', 'зеленый', 'жёлтый']
chosen_color = random.choice(colors)
print(f'Выбранный цвет: {chosen_color}')
```

**Задача 3. Перемешивание списка**
Дан список чисел, перемешать его элементы в случайном порядке и вывести. \
`random.shuffle(sequence)` - перемешивает последовательность (изменяется сама последовательность). Функция не работает для неизменяемых объектов.


```python
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(f'Перемешанный список: {numbers}')
```

**Задача 4. Случайная выборка из списка без повторений**
Дан список имен, выбрать случайно 3 имени без повторений. \
`random.sample(population, k)` - список длиной k из последовательности population.

```python
names = ['Аня','Борис','Вика','Гриша','Дима']
selected_names = random.sample(names, 3)
print(f'Выбранные имена: {selected_names}')
```
