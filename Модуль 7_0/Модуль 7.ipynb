# Модуль random 
Этот модуль используется для генерации случайных чисел и данных. Это может быть полезно для различных сценариев, например, случайный выбор элементов или создание случайных чисел.


**Задача 1. Случайное число в заданном диапазоне** \
Сгенерировать и вывести случайное число от 1 до 10 \
`random.randint(a, b)` - случайное целое число n, a ≤ n ≤ b.

```python
import random
```

```python
for i in range (5):
    num = random.randint(0,100)
    print(f'Случайное число от 1 до 100: {num}')
```

```python
num = random.randint(1, 10)
print(f'Случайное число от 1 до 10: {num}')
```

**Задача 2. Случайный выбор из списка** \
Дан список цветов, случайным образом выбрать и вывести один из них. \
`random.choice(sequence)` - случайный элемент непустой последовательности.

```python
colors = ['красный', 'синий', 'зеленый', 'жёлтый']
chosen_color = random.choice(colors)
print(f'Выбранный цвет: {chosen_color}')
```

**Задача 3. Перемешивание списка**
Дан список чисел, перемешать его элементы в случайном порядке и вывести. \
`random.shuffle(sequence)` - перемешивает последовательность (изменяется сама последовательность). Функция не работает для неизменяемых объектов.


```python
numbers = [1, 2, 3, 4, 5]
random.shuffle(numbers)
print(f'Перемешанный список: {numbers}')
```

**Задача 4. Случайная выборка из списка без повторений**
Дан список имен, выбрать случайно 3 имени без повторений. \
`random.sample(population, k)` - список длиной k из последовательности population.

```python
names = ['Аня','Борис','Вика','Гриша','Дима']
selected_names = random.sample(names, 3)
print(f'Выбранные имена: {selected_names}')
```
