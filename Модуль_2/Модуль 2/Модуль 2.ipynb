# <a id="if">1. Условный оператор if</a>

Оператор ветвления (условная инструкция, условный оператор) — конструкция языка программирования, обеспечивающая выполнение определённой команды только при условии истинности некоторого логического выражения, либо выполнение одной из нескольких команд в зависимости от значения некоторого выражения.
В программировании условный оператор применяется, когда мы хотим описать ход действия программы в разных ситуациях.

Представим такую ситуацию, вам необходимо запросить результат тестирования у пользователя и вывести значение на экран, при этом если результат больше или равен 5 баллов, выходит дополнительно сообщение "Тест успешно пройден".

> **if условие:** \
&nbsp;&nbsp;&nbsp;&nbsp;инструкция \
инструкция

```python
result = 100
if result >= 5:
    print('Тест успешно пройден')
print('Ваш результат' , result)
```

# <a id="if-else">2. Условный оператор if-else</a>

ЕСЛИ (IF) соблюдается какое-то условие — сделать одно, 
В ПРОТИВНОМ СЛУЧАЕ (ELSE) — сделать другое.

> **if условие:** \
&nbsp;&nbsp;&nbsp;&nbsp;инструкция \

&nbsp;&nbsp;&nbsp;&nbsp;инструкция 

Дополним задачу условиями:
- Если пользователь выполняет тест на 5 и более баллов, выходит сообщение "Тест успешно пройден"
- В противном случае - "Попробуйте еще раз"

```python
result = 1
if result >= 5:
    print('Тест успешно пройден')
else:
    print('Попробуйте еще раз')
print('Ваш результат' , result)
```

Python проверяет то, что написано после if, получает вместо всего этого выражения True или False и действует исходя из этого:
- если после if получается True – выполняет инструкции, которые за ним следуют
- если после if получается False – перескакивает на инструкции, стоящие после else

При этом условие в блоке if может содержать как одно логическое выражение, так и несколько выражений, соединённых логическими операторами and или or. 

> **if условие_1 and условие_2** \

Допустим мы собираем заявки на участие в мероприятие. Если было получено от 10 и до 40 заявок выводится фраза "Мероприятие состоится в указанное время", иначе "Мероприятие не состоится"


```python
quantity = 9
if quantity >= 10 and quantity <= 40:
    print('Мероприятие состоится в указанное время')
else:
    print('Мероприятие не состоится')
```

Рассмотрим другой пример. В кафе принимают только напитки: чай или кофе. Если клиент заказал кофе или чай, выводится сообщение, что заказ принят. Если другой напиток – выводится, что заказать этот напиток нельзя.

Вы собрались в отпуск и хотите взять с собой своё домашнее животное. В пассажирский салон самолёта можно брать только собак и кошек, все остальные животные должны лететь в специальном багажном отсеке. Если животное – собака или животное – кошка, выбираем один путь. Во всех других случаях – другой.
Но и в салоне самолёта можно передвигаться по-разному: чьи-то переноски ставят под сиденье пассажира, а другие закрепляют на отдельном кресле (у некоторых авиакомпаний). Это зависит от размера животного. Если вес животного меньше 5 кг, питомец размещается под креслом, иначе в переноске на отдельном кресле.

```python
animal = 'cat'
weight = 1
if animal == 'cat' or animal == 'dog':
    print('You could take your animal with you')
    if weight <= 5:
        print('You animal coud be under your seat')
    else:
        print('Your animal need another seat')
else:
    print('Sorry, your animal will be at baggage')
```

Есть и другой вариант усложнения, когда развилка содержит больше двух путей. Для этого используется оператор elif.

Напишем программу, которая будет с нами здороваться в соответствии со временем суток. Всё, что до 5 утра, будем считать ночью. Всё, что до полудня – утром. До 18 включительно у нас будет день, а всё остальное – вечер.

```python
time = 24
if time < 5:
    print('Доброй ночи')
elif time < 12:
    print('Доброе утро')
elif time <=18:
    print('Добрый день')
else:
    print('Добрый вечер')
```

# <a id="possibilities">4. Дополнительные возможности</a>

Можно использовать значение переменной как False, если она равна 0 и как True во всех других случаях.

Представим ситуацию, что в кафе есть электронный терминал, через который можно сделать заказ. Напишем программу, которая будет выяснять у клиента его платежеспособность. Если есть хоть сколько-то денег, мы предлагаем ему совершить заказ, если нет, отказываем.

```python
money = 0
if money:
    print('Мы рады Вам!')
else:
    print('Ты нищеброт!')
```

Если у человека есть деньги, мы будем вычитать из них какую-то сумму (стоимость блюда) и выводить оставшееся на экран. Для этого перезапишем переменную, вычитая сумму из её предыдущего значения:

```python
money = 15
if money:
    print('Мы рады Вам!')
    print('Все блюда по 5')
    money = money - 5
    print('У Вас осталось', money, 'денег')
else:
    print('Ты нищеброт!')
```

Дополним наш предыдущий пример ещё одним предложением блюда в случае, если у клиента осталось достаточно денег. Добавляем конструкцию с if и проверкой количества денег. Если денег недостаточно, никаких инструкций не даём.

```python
money = 8
if money:
    print('Мы рады Вам!')
    print('Все блюда по 5')
    money = money - 5
    print('У Вас осталось', money, 'денег')
    if money >= 5:
        print('Хотите заказать что-нибудь еще?')
else:
    print('Ты нищеброт!')
```

# <a id="error">5. Ошибки и исключения</a>

Напишем программу, которая поможет нам делить торт на нужное количество гостей. Она будет спрашивать вес торта, количество гостей и выдавать на выходе, сколько в граммах должен получить каждый гость (если делить поровну).

```python
weight = input('Сколько весит торт?')
guests = input('Сколько у Вас гостей?')
result = int(weight) / int(guests)
print('Каждый гость получит по', result, 'граммов торта')
```

Команда try позволяет безопасно попробовать запустить код:
-	Если он запустится без ошибок, он просто выполнится как обычно.
-	Если же возникнет исключение, Python перейдёт к исполнению второй части, которая называется except. В ней можно пояснить ожидаемую ошибку.
-	Последняя часть конструкции – finally – будет выполняться в любом случае, произошла ошибка или нет.

```python
weight = input('Сколько весит торт?')
guests = input('Сколько у Вас гостей?')
try:
    result = int(weight) / int(guests)
    print('Каждый гость получит по', result, 'граммов торта')
except:
    print('Вы неправильно ввели число, мы не можем рссчитать долю каждого гостя')
finally:
    print('Приятного аппетита!')
```
