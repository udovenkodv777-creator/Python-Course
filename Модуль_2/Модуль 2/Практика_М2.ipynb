1.	Напишите программу, которая сообщает размер штрафа в зависимости от превышения скорости. Сейчас действуют такие правила:
-	за превышение скорости от 20 до 40 км/ч предусмотрен штраф 500 рублей,
-	от 40 до 60 км/ч — 1000–1500 рублей,
-	от 60 до 80 км/ч — штраф 2000–2500 рублей,
-	от 80 км/ч — 5000 рублей или лишение прав на полгода. Лишение прав рассматривать не будем.
Пусть ваша программа принимает на ввод разрешённую скорость и скорость, с которой ехал автомобиль, высчитывает разницу и сообщает размер штрафа в зависимости от превышения скорости.


```python
allowed_speed = int(input('Введите разрешенную скорость '))
your_speed = int(input('Введите Вашу скорость '))
dif_speed = your_speed - allowed_speed
if dif_speed in range(20,41):
    print('Штраф 500 руб.')
elif dif_speed in range(40,61):
    print('Штраф 1000-1500 руб.')
elif dif_speed in range(60,81):
    print('Штраф 2000-2500 руб.')
elif dif_speed > 80:
    print('Штраф 5000 руб. или лишение прав на полгода')
else:
    print('Нарушений нет')  
    
    
```

2.	Напишите программу, которая запрашивает на ввод число. После этого она проверяет, является ли число чётным или нечётным. Если число чётное, программа проверяет, делится ли оно на 5 без остатка. Если да, программа сообщает, что вы ввели круглое число. Если нет, сообщает, что число некруглое. Используйте вложенные условия. Для нечётных чисел программа сообщает, что получила нечётное число.

```python
num = int(input('Введите число'))
if num % 2 ==0:
    if num % 5 ==0:
        print('Вы ввели круглое число')
    else:
        print('Число некруглое')
else:
    print('Получено нечетное число')
```

Если указать отрицательное значение индекса, то номер будет отсчитываться с конца, например, s[-1]= "g".

## Трех местное выражение if

```python
a = int(input('Введите число: '))

# Классический вариант
if a >= 0:
    b = a ** 2
else:
    b = abs(a)
    
print(b)
```

```python
a = int(input('Введите число: '))

b = a ** 2 if a >=0 else abs(a)
    
print(b)
```
