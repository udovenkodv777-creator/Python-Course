# <a id="string">1. Строковый метод split() </a>

`.split()` – это заранее определённый ***метод*** для строковых переменных, который делит строку на части по разделительному символу. В результате метода мы получаем список из выделенных частей строк. Применить его можно, написав `split` через точку и добавив круглые скобки. 

Если в круглых скобках ничего не указать, метод по умолчанию считает разделительным символом пробел или последовательность пробелов. 

Попробуем добавить в `fio` пробелов между фамилией и именем и заново применить `.split()`. 

```python
fio.split()
```

Если мы хотим использовать в качестве разделительного символа не пробел, а другой символ, нужно просто указать его внутри скобок:

```python
fio = 'Иванов!Герасим!Александрович'
```

```python
fio = 'Иванов Герасим Александрович'
```

Запишем фамилию, имя  и отчество в соответствующие переменные.

```python
fio = '   Иванов    Герасим    Александрович'
slitted_fio = fio.split()
surname = slitted_fio[0]
name = slitted_fio[1]
middle_name = slitted_fio[2]

print(surname)
print(name)
print(middle_name)
```

```python
fio.split()
```

# <a id="fstring">2. f-строки </a>

Вернёмся к ФИО, которые мы извлекли из строки. Теперь представим, что нам нужно сделать красивый форматированный вывод этих данных на экран в виде:

`Фамилия:Иванов,Имя:Герасим,Отчество:Александрович`

Мы уже изучили два метода вывода на экран нескольких переменных сразу:

1. Указать все значения через запятую:

```python
print('Фамилия:', surname,';', 'Имя:', name,';','Отчество:',middle_name)
```

Указывая значения через запятую, Python автоматически проставляет пробелы между ними. 

2. Конкатенация: 

```python
print('Фамилия:' + surname + '; Имя:' + name + '; Отчество:' + middle_name)
```

3. f-строки:

```python
print(f'Фамилия: {surname}; Имя: {name}; Отчество: {middle_name}')
```

Мы воспользовались ***f-строками***. f – значит форматированная строка. Такие строки внутри фигурных скобок могут подставлять значения из переменных, которые объявлены в текущем файле области видимости, могут исполнять любые арифметические действия, и даже вызывать функции. 

Подставим арифметическое выражение вместо фамилии:

```python
print(f'Фамилия: {2 + 3}; Имя: {name}; Отчество: {middle_name}')
```

Вызовем функцию `len()` от строки `fio`:

```python
b = 3.141233
print(f'число b = {b:.2f}')
```

```python
a = 2.79999
print(f'{a:.2f}')
```

С помощью f-строк можно так же форматировать числа. \
Возьмем переменную a и присвоим ей значение 2.000000001, выведем на экран результат с 2-умя знаками после запятой.

```python
a = 2.7999999
```

# <a id="replace_join">4. Методы replace() и join() </a>

Представим, что нужно автоматизировать процесс полученных домашних заданий. Имя файла, которые вы получили от студента ‘ДЗ 4’, необходимо преобразовать в стандартизированную запись, которая включает 'ФИО студента. Домашнее задание номер. Дата'

```python
#исходные данные
fio = 'Иванов Герасим Александрович'
date = '2025-10-29'
file_name = 'ДЗ 4'
```

1. Очистка и замена (`.replace()`) 
Заменим значение 'ДЗ 4' на полное название 'Домашнее задание 4', используя метод replace()

`str.replace()` — или просто replace() — заменяет одну часть строки на другую.

Синтаксис метода str.replace(старый_текст, новый_текст, количество). Укажите, что именно нужно заменить, на что и сколько раз. 

```python
new_file_name = file_name.replace('ДЗ','Домашнее задание')
print(f'Очищенное имя: {new_file_name}')
```

2. Сбока данных для записи (подготовка списка)

```python
full_name = [fio, new_file_name, date]
print(full_name)
```

3. Объединяем в полное название с помощью метода `join()` \
Этот метод применяется к строкам и позволяет объединить элементы списка в одну строку с указанным разделителем.
`''.join()`, где \
`''` — это разделитель, который мы используем \
`()` — сюда подставляется аргумент: список элементов, которые необходимо объединить

```python
result_file_name = '. '.join(full_name)
print(result_file_name)
```

# <a id="normalize">5. Нормализация строк </a>

Нормализация – это процесс преобразования строк таким образом, чтобы они были сопоставимы и легко обрабатывались, независимо от того, как они были введены изначально.

Нормализация Регистра: `.lower()` и `.upper()`

Метод: `.lower()` (приведение к нижнему регистру) \
Этот метод переводит **все** буквенные символы в строке в нижний регистр.

Рассмотрим пример: проверка пароля без учета регистра (*для реальных паролей это не рекомендуется)

```python
user_input = 'adMin12345'
password = 'admin12345'
```

```python
#без нормализации
print(user_input == password)
```

```python
#с нормализацией
print(user_input.lower() == password)
```

Метод: `.upper()` (приведение к верхнему регистру)

Этот метод переводит все буквенные символы в строке в верхний регистр.

**Пример:** Приведение кода страны к единому формату.

```python
country_code = 'ru'
currence_code = 'usd'
```

```python
print(f'Код страны: {country_code.upper()}')
print(f'Курс валюты: {currence_code.upper()}')
```

`.strip()` с Аргументом (удаление специфических символов)

`Метод lstrip()`
Если strip() чистит всё, то lstrip() (от "left strip") убирает символы только слева.

```python
code = '---ABS-001****'
print(code.lstrip('-'))
```

`Метод rstrip()`
Аналогично rstrip() ("right strip") чистит только справа.

```python
code = '---ABS-001****'
print(code.rstrip('*'))
```
