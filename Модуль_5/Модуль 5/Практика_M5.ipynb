# Практика. Обработка текстовой информации

1. Дана строка чисел, разделенных пробелом. Сформировать список чисел и возвести в квадрат каждый элемент списка. Результат вывести в виде строки чисел.

```python
numbers = [1, 2, 3, 4, 5]

# Ваш код
squares = [str(x ** 2) for x in numbers]
new_numbers = ' '.join(squares)

print(new_numbers)
```

2. Сформируйте список из предприятий 'ООО' и выведите их на экран. 

```python
contr = 'ЗМКООО,ООО;ЗМКООО-1,ООО;ТОЛЬЯТТИСИНТЕЗ,АО;ТЕХНОСЕРВИС УЦ ДПО,АНО;ПРОМТРАНСЭНЕРГО,ООО;АВДН-СТРОЙ,ООО'

# Разбиваем строку на список предприятий по точке с запятой
companies = contr.split(';')

# Фильтруем список, оставляя только предприятия с 'ООО'
ooo_list = [company for company in companies if 'ООО' in company]

# Выводим результат на экран
print(ooo_list)
```

3. Дан список ФИО и электронных адресов участников встречи, найдите тех, у кого не указаны электронные адреса.

```python
participant = ''' Лубнов Михаил Юрьевич <LubnovMYur@alfa.com>;  Палкин  Илья Федорович <PalkinIF@alfa.com>; 
 Шалкина Татьяна Николаевна <ShalkinaTN@alfa.com>; Козлова Екатерина Викторовна; 
Звонкова Анна Семеновна <ZvonkovaAS@alfa.com>; Лаврова Валерия Алексеевна <LavrovaVA@alfa.com>; 
Кучкина Екатерина Сергеевна; Николаева Екатерина Арнольдовна <NikolaevaEA@alfa.com>
'''

# Разбиваем строку по точке с запятой и очищаем от пробелов по краям
people = [p.strip() for p in participant.split(';') if p.strip()]

# Фильтруем тех, у кого в строке нет символа '@'
no_email = [p for p in people if '@' not in p]

# Выводим результат
for person in no_email:
    print(person)
```
