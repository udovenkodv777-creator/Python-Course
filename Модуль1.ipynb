# <a id="types">2. Типы данных </a>
Тип данных позволяет определить программе, каким способом можно обрабатывать данные.
Выделяют:
- числовой тип 
- строковый тип (последовательность символов в кавычках)
- логический тип (True или False)

### Числа делятся на:
- целые (integer / int)
- дробные (float)
С числами можно производить арифметические операции. В одном выражении можно использовать и целые, и дробные числа. Иногда результат будет целым числом, иногда — дробным.

Если сложить два целых числа, на выходе также получится целое:

```python
10 + 5
```

Если сложить два дробных, получится дробное:

```python
1.5 + 1.5
```

Если целое и дробное — получится дробное:

```python
2 + 1.5
```

У дробного всегда будет указана дробная часть, даже если она равна нулю:

```python
7.4 + 6.6
```

Чтобы узнать, к какому типу данных относится интересующее нас число, нужно использовать функцию type():

```python
type(3)
```

```python
type(1.4)
```

```python
type(6.0)
```

### Приведение типов

```python
float(7)
```

```python
int(4.9)
```

```python
round(4.9)
```

## Правила имен переменных
Следует придерживаться нескольких несложных правил при выборе имен переменных:

1. Имя переменной может содержать только латинские буквы (от a до Z), числа (от 0 до 9) и символ нижнего подчеркивания(_);
2. Имя переменной не должно содержать пробелов;
3. Имя переменной не должно начинаться с цифры;
4. Регистр важен: var и Var это разные переменные.

 Имена переменных не должны совпадать с ключевыми словами и встроенными функциями(пр. def, False)
 Негласная договоренность об именах: не использовать одиночные буквы l, O, или I в качестве каких-либо имен из-за риска спутать их с 1 и 0, в зависимости от шрифта. Если вам хочется использовать «l», используйте «L».

# <a id="boolean">6. Булевы значения </a>

В Python булевы значения — это логический данных, который принимает только два значения: `True` и `False`. Правда или ложь.

С помощью них можно проверить, верно ли какое-то утверждение. Например, верно ли, что 3 больше 5? Для сравнения мы используем операторы сравнения `>` и `<`:

```python
3 > 5
```

```python
3 < 5
```

```python
100 > 10000
```

Похожим образом можно проверить равенство и неравенство значений. Для сравнения значений используется двойной знак равенства ==:

```python
35 == 120
```

Знак неравенства записывается как восклицательный знак и равно: !=

```python
25 != 120
```

Можно также использовать операторы >= ("больше либо равно") и <= ("меньше либо равно"):

```python
3 <= 5
```

```python
35 >= 35
```

Строки тоже можно сравнивать. В данном случае оператор == будет значить, что строки полностью совпадают:

```python
'пятница' == "пятница"
```

Кроме операторов сравнения в Python существуют также логические операторы and, or и not.
Оператор and (И) будет выдавать в выводе True, только если верны оба утверждения:

```python
(3 > 5) and (15 > 10)
```

```python
(3 > 5) or (15 > 10)
```

```python
not 35 > 30
```

```python
not 99 == 100
```

Оператор `not` (НЕ) возвращает `True`, если утверждение неверно:

Булевы значения также используются в функциях: некоторые функции в качестве результата возвращают значения `True` или `False`.

Например, функция `isinstance()`. Мы передаем функции какой-то объект и название типа данных. Если этот объект относится к этому типу данных — функция вернёт `True`. Если объект другого типа — вернёт `False`.

```python
isinstance(25, int)
```

```python
isinstance(25.5, int)
```

```python
isinstance('25.5', int)
```
