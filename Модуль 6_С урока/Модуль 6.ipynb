# <a id="function">1. Работа с функциями </a>

**Функция** – это именованный блок кода, который выполняет определённую задачу. 

*Именованный* значит, что у него есть некоторое имя, по которому можно обращаться к этому коду. 

Один раз объявив функцию, мы сможем вызывать её по имени в любом месте нашей программы.



Функции бывают с входными параметрами и без. 
Функции, которые на вход ничего не принимают, называются *функциями без входных параметров*. 

### Объявление функции в Python

Правила для имен функций: 
- Имена функций всегда пишутся в нижнем регистре
- Слова разделяются нижним подчеркиванием
- Имена не могут начинаться со знаков препинания или цифр 
- Имена должны быть говорящими – то есть транслировать то, что делает функция 
- Функции нельзя называть зарезервированными в Python именами. Список  зарезервированных имён можно увидеть, вызвав команду `help("keywords")`:

### Задача:
Нужно посчитать метрику NPS (Net Promoter Score – индекс потребительской лояльности) по сервису в трёх регионах и по всем регионам в целом. Данные по пользовательским оценкам представлены в виде списка списков. 

```python
scores_list = [
                    [10, 10, 9, 10, 1, 8, 5],
                    [9, 9, 8, 5],
                    [10, 10, 10, 4, 5]
]
```

Пользователи оценивают готовность рекомендовать сервис по шкале от 1 до 10:
- 9 и 10 – это самые лояльные пользователи (positive), 
- 1-6 – критики (negative), 
- 7-8 – нейтральные пользователи (neutral). 

Для расчёта используется формула NPS = % (positive) - % (negative)

```python
def nps (scores):
    positive_score = 0
    negative_score = 0
    
    for score in scores:
        if score >= 9:
            positive_score += 1
        elif score <= 6:
            negative_score += 1
            
    nps = (positive_score - negative_score)/len(scores)
    nps = nps * 100
    
    return nps
```

```python
scores_1 = [10, 10, 9, 10, 1, 8, 5]
nps(scores_1)
```

```python
def extend_list(scores_list):
    result = []
    for score in scores_list:
        #result += score
        result.extend(score)
    return result
```

```python
def total_nps (scores_list):
    result_list = extend_list(scores_list)
    return nps(result_list)
```

```python
for score in scores_list:
    print(f'NPS региона {nps(score)}')

print(f'Общий nps {total_nps(scores_list)}')
```

# <a id="recursion">4. Рекурсия </a>

**Рекурсивная функция** — это функция, которая вызывает сама себя напрямую или через другие функции. 

Числа Фибоначчи \
Числа Фибоначчи — это числовая последовательность, где каждое следующее число является суммой двух предыдущих, начиная с 0 и 1. \
`f` — функция расчёта числа Фибоначчи, `n` — это номер числа Фибоначчи в последовательности, `f(n)` — это число Фибоначчи, находящееся в последовательности под номером `n`. \
`f(n) = f(n-1) + f(n-2), n>=2` \
Поскольку первый и нулевой элементы последовательности мы по этой формуле рассчитать не можем, явно запишем, что они равны 1: \
`f(1) = 1, f(0) = 1`


```python
def fibonacci (n):
    if n == 0 or n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
```
